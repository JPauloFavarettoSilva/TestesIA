```python
import gensim.downloader as api


glove_model = api.load("glove-wiki-gigaword-100")


word_vector = glove_model['apple']
print("Vetor GloVe para a palavra 'apple':", word_vector)

similar_words = glove_model.most_similar('apple', topn=5)
print("Palavras mais similares a 'apple':", similar_words)


word_vector = glove_model['ai']
print("Vetor GloVe para a palavra 'Ai':", word_vector)

similar_words = glove_model.most_similar('ai', topn=5)
print("Palavras mais similares a 'Ai':", similar_words)

word_vector = glove_model['engine']
print("Vetor GloVe para a palavra 'Engine':", word_vector)

similar_words = glove_model.most_similar('engine', topn=5)
print("Palavras mais similares a 'Engine':", similar_words)

word_vector = glove_model['food']
print("Vetor GloVe para a palavra 'Food':", word_vector)

similar_words = glove_model.most_similar('food', topn=5)
print("Palavras mais similares a 'Food':", similar_words)

word_vector = glove_model['think']
print("Vetor GloVe para a palavra 'Think':", word_vector)

similar_words = glove_model.most_similar('think', topn=5)
print("Palavras mais similares a 'think':", similar_words)
```
